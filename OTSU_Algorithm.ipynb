```python
import numpy as np
import matplotlib.pyplot as plt
import skimage
from skimage import io
from skimage import viewer
import cv2
from google.colab.patches import cv2_imshow
from skimage.viewer import ImageViewer

def histogram(img):
  # I/p the image
  image = cv2.imread(img)
  image = cv2.resize(image, (400, 500))
  
  #From RGB to grayscale
  gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  
  #Display the Image
  print("      \t      Orignal Image")
  cv2_imshow(image)
  print ("      \t   Grey Scale Image")
  cv2_imshow(gray)

  #Display the histogram
  print ("     \t   Histogram of the grayscale Image")
  ax = plt.hist(gray.ravel(), bins = 256)
  plt.show()
   
def binarization(img,threshold):
  # I/p the image
  image = cv2.imread(img)
  
  #From RGB to grayscale
  gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

  #Perform thresholding
  output = gray < threshold
  skimage.io.imshow(output)

def otsu(img):
    
    # I/p the image
    image = cv2.imread(img)

    # From rgb to greyscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  

    #Get image histogram
    hist, edges = np.histogram(gray, bins=256)

    #Normalization
    hist = np.divide(hist.ravel(), hist.max())
    mids = (edges[:-1] + edges[1:]) / 2.

    # To calculate probabilities p1, p2
    p1 = np.cumsum(hist)
    p2 = np.cumsum(hist[::-1])[::-1]

    #Calculating mean
    u1 = np.cumsum(hist * mids) / p1
    u2 = (np.cumsum((hist * mids)[::-1]) / p2[::-1])[::-1]

    # Calculation
    sig_square = p1[:-1] * p2[1:] * (u1[:-1] - u2[1:]) ** 2
    index_max = np.argmax(sig_square)

    threshold = mids[:-1][index_max]

    print("Threshold value calculated by OTSU algorithm  is : ", threshold)

    #Thresholding
    Output = gray < threshold

    #Result
    skimage.io.imshow(Output)
    


```
